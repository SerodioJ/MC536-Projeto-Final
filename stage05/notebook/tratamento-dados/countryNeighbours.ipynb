```python
import pandas as pd
import numpy as np
```

```python
countries = pd.read_csv('../../data/processed/countries.csv')
neighbours = pd.read_csv('../../data/external/CountriesNeighbourhood.csv')
```

```python
countryList = countries['country'].tolist()
labelList = countries['label'].tolist()
countryList2 = neighbours['country'].tolist()
neighboursList = neighbours['neighbour'].tolist()
```

```python
rows = []
dic = {
    'Republic of the Congo': 'Congo',
    'Czech Republic': 'Czechia',
    'Russia': 'Russian Federation',
    "People's Republic of China": 'China',
    'Tanzania': 'United Republic of Tanzania',
    'Laos': "Lao People's Democratic Republic",
    'Vietnam': 'Viet Nam',
    'United Kingdom': 'United Kingdom of Great Britain and Northern Ireland',
    'East Timor': 'Timor-Leste',
    'Syria': 'Syrian Arab Republic',
    'United States': 'United States of America',
    'Swaziland': 'Eswatini',
    'Moldova': 'Republic of Moldova',
    'Brunei': 'Brunei Darussalam',
    'Cape Verde': 'Cabo Verde',
    'North Korea': "Democratic People's Republic of Korea",
    'The Gambia': 'Gambia',
    'South Korea': 'Republic of Korea'
}

ignore = ['Liechtenstein', 'Palestine', 'Andorra', 'Monaco', 'Iraq', 'San Marino', 'Vatican City']
```

```python
for country, label in zip(countryList, labelList):
    print(country, label)
    try:
        index = countryList2.index(list(dic.keys())[list(dic.values()).index(country)]) if country in dic.values() else countryList2.index(country)
    except:
        index = -1
    while (index!=-1):
        if isinstance(neighboursList[index], str) and neighboursList[index] not in ignore:
            rows.append({'country': label, 'neighbour': labelList[countryList.index(dic[neighboursList[index]])] if neighboursList[index] in dic else labelList[countryList.index(neighboursList[index])]})
        try:
            index = countryList2.index(list(dic.keys())[list(dic.values()).index(country)], index+1) if country in dic.values() else countryList2.index(country, index+1)
        except:
            index = -1
neighbors = pd.DataFrame(rows, columns=['country', 'neighbour'])
```

```python
neighbors.to_csv('../../data/processed/neighbours.csv', index=False)
```
