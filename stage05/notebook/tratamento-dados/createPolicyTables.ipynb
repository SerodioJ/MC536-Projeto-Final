Bibliotecas

```python
import pandas as pd
import numpy as np
```

### Criação da tabela de políticas

```python
rows = []
tabelas = ['legislacao', 'tratamento', 'testagem', 'prevencao']
dict_type = {'legislacao': 'law', 'tratamento': 'treatment', 'testagem': 'testing', 'prevencao': 'prevention'}
id_tracker = 1
for tabela in tabelas:
    csv = pd.read_csv("../../data/processed/{}.csv".format(tabela))
    for column in csv.columns[1:]:
        rows.append({'id': 'P{}'.format(id_tracker), 'policy': column.strip(), 'type': dict_type[tabela]})
        id_tracker += 1
policies = pd.DataFrame(rows, columns=['id', 'policy', 'type'])
```

```python
policies.to_csv('../../data/processed/policies.csv', index=False)
```

### Criação da tabela de políticas de cada país

```python
rows = []
tabelas = ['legislacao', 'tratamento', 'testagem', 'prevencao']
dict_type = {'legislacao': 'law', 'tratamento': 'treatment', 'testagem': 'testing', 'prevencao': 'prevention'}
for tabela in tabelas:
    csv = pd.read_csv("../../data/processed/{}.csv".format(tabela))
    columns = csv.columns[1:]
    for i in range(len(csv)):
        row = csv.loc[i]
        country = row['Country']
        for column in columns:
            value = row[column]
            if value == '':
                continue
            isNumeric = isinstance(value, int) or isinstance(value, float)
            if isNumeric and np.isnan(value):
                continue
            rows.append({'country': country.strip(), 
                         'policyId': policies.loc[(policies['policy'] == column.strip()) & (policies['type'] == dict_type[tabela])]['id'].values[0], 
                         'isNumeric': isNumeric,
                         'value': '' if isNumeric else value.strip(),
                         'numericValue': int(value) if isNumeric else ''})
        print('{} - {}/{}'.format(tabela, i+1, len(csv)))
country_policies = pd.DataFrame(rows, columns=['country', 'policyId', 'isNumeric', 'value', 'numericValue'])
```

```python
country_policies.to_csv('../../data/processed/country_policies.csv', index=False)
```
