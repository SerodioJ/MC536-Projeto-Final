# Politica de distribuicao de camisinhas

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
camisinha = pd.read_csv('../../saida/distribuicao-regiao.csv')
camisinha
```

```python
camisinha.rename(columns = {'Tem uma política ou estratégia nacional de distribuição de camisinhas?' : 'politica'}, inplace = True)
```

```python
camisinha = camisinha.replace(['Western Pacific'],'West-Pac')
camisinha = camisinha.replace(['Eastern Mediterranean'], 'East-Med')
camisinha = camisinha.replace(['South-East Asia'], 'Asia')
camisinha
```

```python
temPol = camisinha[camisinha['politica'] == 'Yes']
naoTemPol = camisinha[camisinha['politica'] == 'No']
```

```python
camisinha['REGIAO'][0]
```

```python
plt.bar(temPol['REGIAO'], temPol['QUANTIDADE'], color ='y', edgecolor ='blue', label ='Sim')
plt.bar(naoTemPol['REGIAO'], naoTemPol['QUANTIDADE'], color ='r',  edgecolor ='blue', label ='Nao')
plt.xlabel('Regiao')
plt.ylabel('Numero de paises')
plt.title('Contagem de paises com distribuicao de camisinhas em 2014')
plt.savefig('../../saida/camisinha.png')
```

# Regiao vs Infeccao

```python
regiao = pd.read_csv('../../saida/regiao-x-infeccoes.csv')
regiao
```

```python
regiao2 = regiao[ regiao['ANO'] == 2014 ]
regiao2
```

```python
regiao2 = regiao2.replace(['Western Pacific'],'West-Pac')
regiao2 = regiao2.replace(['Eastern Mediterranean'], 'East-Med')
regiao2 = regiao2.replace(['South-East Asia'], 'Asia')
regiao2
```

```python
plt.bar(regiao2['REGIAO'], regiao2['INFECCOES'], color ='y', edgecolor ='blue', label ='Sim')
plt.xlabel('Regiao')
plt.ylabel('% infeccoes em relacao ao total mundial')
plt.title('Quantidade de infeccoes por HIV em relacao ao valor mundial por regiao ')
plt.savefig('../../saida/regiao.png')
```
