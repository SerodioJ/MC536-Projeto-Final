```
import json
import requests
```

```
def getValue(dado):
  return {'min': dado['value']['low'],
          'max': dado['value']['high'],
          'nominal': dado['value']['numeric']}
```

```
# Get list of countries used in WHO dataset
def getCountriesList():
  countriesList = []
  url = 'https://apps.who.int/gho/athena/api/GHO/HIV_0000000026?format=json'
  js = requests.get(url).json()
  for i in range(170):  
    countriesList.append(js['dimension'][4]['code'][i]['label'])
  return countriesList
```

```
url = 'https://apps.who.int/gho/athena/api/GHO/HIV_0000000026?format=json'
arquivo = requests.get(url).json()
```

```
countriesList = getCountriesList()
globalListData = []
allDataSTI = []
```

```
print(countriesList, len(countriesList))
```

```
countriesListData = []          #List with data about countries' infections 
for i in range(len(countriesList)):
  countriesListData.append([])
```

```
errorCount = 0
for dado in arquivo['fact']:
  year = 0
  region = ''
  country = ''

  if dado['published']:
    for i in dado['Dim']: # Find data about the country, region and name
      if i['category'] == 'REGION':
        region = i['code']
      elif i['category'] == 'YEAR':
        year = int(i['code'])
      elif i['category'] == 'COUNTRY':
        country = i['code']
    if region == 'GLOBAL':
      globalListData.append({'year': year,
                             'value': getValue(dado)})
      nInfections = getValue(dado)
      allDataSTI.append({'region': 'GLOBAL',
                         'year': year,
                         'min': nInfections['min'],
                         'nominal': nInfections['nominal'],
                         'max': nInfections['max']})
    else:
      try:
        nInfections = getValue(dado)
        allDataSTI.append({'region': country,
                           'year': year,
                           'min': nInfections['min'],
                           'nominal': nInfections['nominal'],
                           'max': nInfections['max']})
        index = countriesList.index(country)
        countriesListData[index].append({'year': year,
                                         'value': getValue(dado)})
      except:
        errorCount += 1
        print('\"%s\" not found, %d'%(country,errorCount))
```

```
print(globalListData)
```

```
import pandas as pd

df = pd.DataFrame(allDataSTI,columns=['region', 'year', 'min', 'nominal', 'max'])
```
